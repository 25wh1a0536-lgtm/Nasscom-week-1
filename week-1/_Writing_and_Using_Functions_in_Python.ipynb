```python
def greet():
    print("Hello! Welcome to Python Functions.")

greet()
```

```python
def greet_user(name):
    print(f"Hello, {name}!")

greet_user("Sreemai")
greet_user("mai")
```

```python
def greet_user(name="Guest"):
    print(f"Welcome, {name}!")

greet_user()
greet_user("Sreemai")
```

```python
def print_names(*names):
    print("Names passed:", names)
    for name in names:
        print("Hello", name)

print_names("Sreemai", "Pranavi", "Unnathi")
```

```python
def print_info(**info):
    print("User Info:")
    for key, value in info.items():
        print(f"{key} = {value}")

print_info(name="Sreemai", age=18, city="Hyderabad")
```

```python
def show_details(*args, **kwargs):
    print("Args:", args)
    print("Kwargs:", kwargs)

show_details(1, 2, 3, name="Sreemai", city="Hyderabad")
```

```python
def multiply(a, b):
    return a * b

def square(n):
    return multiply(n, n)

print("Square of 4:", square(4))
```
