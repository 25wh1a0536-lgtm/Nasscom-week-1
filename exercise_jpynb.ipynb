```python
import os
import datetime
import random


DEVICE_TYPES = [
    'Temperature_Sensor',
    'Humidity_Sensor',
    'Pressure_Sensor',
    'Light_Sensor'
]

LOCATIONS = [
    'Lab1',
    'Lab2',
    'OfficeA',
    'OfficeB',
    'Warehouse'
]

READING_RANGES = {
    'Temperature_Sensor': (18.0, 30.0),
    'Humidity_Sensor': (30.0, 80.0),
    'Pressure_Sensor': (980.0, 1030.0),
    'Light_Sensor': (100, 1000)
}


raw_sensor_data = []


n = int(input("Enter number of sensor readings: "))

for i in range(1, n + 1):


    timestamp_str = datetime.datetime.now().strftime(
        "%Y-%m-%d %H:%M:%S"
    )


    device_id = f"DEV{i:03d}"


    device_type = random.choice(DEVICE_TYPES)


    location = random.choice(LOCATIONS)


    low, high = READING_RANGES[device_type]
    value = round(random.uniform(low, high), 2)

    reading = (
        timestamp_str,
        device_id,
        device_type,
        location,
        value
    )

    raw_sensor_data.append(reading)


print("\nSensor Readings:\n")

for item in raw_sensor_data:
    print(item)
 # 1.2 Create temperature_readings list

temperature_readings = []

for reading in raw_sensor_data:

    timestamp, device_id, device_type, location, value = reading

    if device_type == "Temperature_Sensor":

        temp_dict = {
            "timestamp": timestamp,
            "device_id": device_id,
            "location": location,
            "temperature_celsius": value
        }

        temperature_readings.append(temp_dict)

print("\nTemperature Readings:\n")

for item in temperature_readings:
    print(item)


# 1.3 Create unique_locations set

unique_locations = set()

for reading in raw_sensor_data:

    unique_locations.add(reading[3])

print("\nUnique Locations:\n")
print(unique_locations)


# 1.4 Create device_stats dictionary

device_stats = {}



for device in DEVICE_TYPES:

    device_stats[device] = (float('inf'), float('-inf'))



for reading in raw_sensor_data:

    device_type = reading[2]
    value = reading[4]

    min_value, max_value = device_stats[device_type]

    if value < min_value:
        min_value = value

    if value > max_value:
        max_value = value

    device_stats[device_type] = (min_value, max_value)

print("\nDevice Statistics:\n")

for device, stats in device_stats.items():

    print(device, "->", stats)
print("\n--- Task 1.5: Advanced Indexing and Slicing ---")

sample_string = "PythonProgrammingIsFun"

# 1.5.1 List Slicing



sliced_readings_1 = raw_sensor_data[9:20]

print("\n10th to 20th readings:\n")

for item in sliced_readings_1:
    print(item)



sliced_readings_2 = raw_sensor_data[::5]

print("\nEvery 5th reading:\n")

for item in sliced_readings_2:
    print(item)


# 1.5.2 String Slicing


substring_1 = sample_string[6:17]

print("\nSubstring 1:", substring_1)


substring_2 = sample_string[5::-1]

print("Substring 2:", substring_2)
# 1.5.3 Mixed Slicing



fifth_reading_device_type = raw_sensor_data[4][2]

print("\nDevice type of 5th reading:",
      fifth_reading_device_type)



device_type_suffix = fifth_reading_device_type[-6:]

print("Suffix:", device_type_suffix)
print("\n--- Task 2: Function Definitions and Usage ---")


# 2.1 Function to get readings by location

def get_readings_by_location(readings_list, target_location):

    result = []

    for reading in readings_list:

        if reading[3] == target_location:
            result.append(reading)

    return result




location = input("\nEnter location: ")

location_readings = get_readings_by_location(
    raw_sensor_data,
    location
)

print("\nReadings for", location)

for item in location_readings:
    print(item)
# 2.2 Function to calculate average temperature

def calculate_average_temperature(temp_data_list):

    if len(temp_data_list) == 0:
        return 0.0

    total = 0

    for data in temp_data_list:

        total += data["temperature_celsius"]

    average = total / len(temp_data_list)

    return round(average, 2)


avg_temp = calculate_average_temperature(
    temperature_readings
)

print("\nAverage Temperature =", avg_temp, "°C")



# 2.3 Function to generate report line

def generate_report_line(reading_dict):

    report = (
        f"[{reading_dict['timestamp']}] "
        f"Device {reading_dict['device_id']} "
        f"at {reading_dict['location']}: "
        f"{reading_dict['temperature_celsius']}°C"
    )

    return report


print("\nReport Lines:\n")

for reading in temperature_readings:

    print(generate_report_line(reading))
print("\n--- Task 3: Core Library Usage ---")

# 3.1 Create directory

reports_dir = "sensor_reports"

if not os.path.exists(reports_dir):

    os.makedirs(reports_dir)

    print(f"\nDirectory '{reports_dir}' created.")

else:

    print(f"\nDirectory '{reports_dir}' already exists.")


# 3.2 Calculate duration

if len(raw_sensor_data) > 1:

    first_timestamp_str = raw_sensor_data[0][0]

    last_timestamp_str = raw_sensor_data[-1][0]

    date_format = "%Y-%m-%d %H:%M:%S"

    first_dt = datetime.datetime.strptime(
        first_timestamp_str,
        date_format
    )

    last_dt = datetime.datetime.strptime(
        last_timestamp_str,
        date_format
    )

    duration = last_dt - first_dt

    days = duration.days
    seconds = duration.total_seconds()

    hours = int(seconds // 3600) % 24

    minutes = int((seconds % 3600) // 60)

    print(
        f"\nDuration of data collection: "
        f"{days} days, {hours} hours, {minutes} minutes."
    )


# 3.3 Generate random numbers

random_values = []

for i in range(5):

    value = random.randint(100, 200)

    random_values.append(value)

print("\nRandom Values:")

print(random_values)




```
